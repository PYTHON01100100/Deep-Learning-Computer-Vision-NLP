```python
import pandas as pd
```

```python
data = pd.read_csv(r'/content/drive/MyDrive/Data/Course/Corona.csv',header = 0, encoding = 'cp437')
data
```

```python
data = data[::10]
data.shape
```

```python
data['C'] = data['OriginalTweet'].apply(lambda x : len(x))
data['S'] = data['OriginalTweet'].apply(lambda x : len(x.split()))
for i in ';!@#$%^&*' :
  data[i] = data['OriginalTweet'].apply(lambda x : x.count(i) )
data
```

```python
data.columns
```

```python
X = data[['C', 'S', ';', '!', '@', '#', '$', '%', '^', '&', '*']]
X
```

```python
from sklearn.cluster import KMeans
```

```python
KMeansModel = KMeans(n_clusters=3,init='k-means++', #also can be random
                     random_state=33,algorithm= 'auto') # also can be full or elkan
```

```python
KMeansModel.fit(X)
```

```python
print('KMeansModel centers are : ' , KMeansModel.cluster_centers_)
```

```python
pd.DataFrame(KMeansModel.cluster_centers_,columns = X.columns)
```

```python
KMeansModel = KMeans(n_clusters=5,init='k-means++', #also can be random
                     random_state=33,algorithm= 'auto') # also can be full or elkan
```

```python
KMeansModel.fit(X)
```

```python
print('KMeansModel centers are : ' , KMeansModel.cluster_centers_)
```

```python
pd.DataFrame(KMeansModel.cluster_centers_,columns = X.columns)
```

```python
print('KMeansModel labels are : ' , KMeansModel.labels_)
```

```python
print('KMeansModel intertia is : ' , KMeansModel.inertia_)
```

```python
print('KMeansModel No. of iteration is : ' , KMeansModel.n_iter_)
```

```python
SSDDict = {}
for C in range(2,21):
  KMeansModel = KMeans(n_clusters=C,init='k-means++', random_state=33,algorithm= 'auto')
  KMeansModel.fit(X)
  SSDDict[C] = KMeansModel.inertia_
```

```python
SSDDict = {i:k/1000000 for i,k in zip(SSDDict.keys(),SSDDict.values())}
SSDDict
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (20, 12))
plt.title('SSD for KMeans')
plt.xlabel('Number of Clustering')
plt.ylabel('SSD')

XGraph =SSDDict.keys()
yGraph = SSDDict.values()

#----------------------------------------------------
# Drawing plot graph No 1 . .  :
plt.plot(XGraph,yGraph,linewidth=1,color= 'r', # can be : b ,g ,y ,c ,m ,k ,w
         alpha=1,linestyle = 'solid') # can be  : dotted, dashed , dashdot




plt.show()
```
